```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sy
%matplotlib inline
```

```python
#number of terms to use in sum
N = 10000
#number of time direcretized intervals
m = 200
#terminal time assuming t_0 = 0
t = 10
#time sub-interval
dt = t/m
#positions
X = np.zeros(m+1)
Y = np.zeros(m+1)

def iterate1(x, dt, N):
    noise = (dt/np.sqrt(np.pi))*np.random.normal(0,1)
    #not efficient b/c loop, could do sum of array, try making iterate2
    for i in range(1,N):
        noise = noise + np.sqrt(2/np.pi)*(np.sin(i*dt)/i)*np.random.normal(0,1)
    return x + noise

T = np.linspace(0,t,m+1)
for j in range(1,m+1):
    X[j] = iterate1(X[j-1],dt,N)
    Y[j] = iterate1(Y[j-1],dt,N)
```

```python

plt.plot(T,X)
plt.plot(T,Y)
plt.show()
```
