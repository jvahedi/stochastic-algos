```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sy
%matplotlib inline
```

```python
#function
def g1(x):
    return x**2/s(x)

def g2(x):
    return np.exp(x)/s(x)
    
```

```python
A1 = np.empty(100)
i = 0
for n in range(10,1010,10):
    X = np.random.rand(n)
    G = g1(X)
    A1[i] =np.mean(G)
    i = i+1

A2 = np.empty(100)
i = 0
for n in range(10,1010,10):
    X = np.random.rand(n)
    G = g2(X)
    A2[i] =np.mean(G)
    i = i+1
```

```python
N = np.linspace(10,1000,100)
plt.plot(N,A1)
plt.plot(N,np.ones(100)*(1/3))
plt.show()

plt.plot(N,A2)
plt.plot(N,np.ones(100)*(np.exp(1)-1))
plt.show()
```

```python
#intialization
size = 10**7
X = np.empty(size)
x = .5
sig = .01

def s(u):
    return 1+.1*np.sin(2*np.pi*u)

for i in range(size):
    #iteration
    #-Step 1 (Proposal)
    prop = np.random.rand()
    #-Step 2 (Acceptance Probabilty)
    a = (s(prop))/(s(x))
    a = np.amin([1,a])
    #-Step 3 (Decision)
    r = np.random.rand()
    if r<a:
        x = prop
    #store value
    X[i] = x
    
R = X[int(size/2):]
```

```python
A3 = np.empty(100)
i = 0
for n in range(10,1010,10):
    G = g1(R[:n])
    A3[i] =np.mean(G)
    i = i+1
```

```python
N = np.linspace(10,1000,100)
plt.plot(N,A3)
plt.plot(N,np.ones(100)*(1/3))
plt.show()
```

```python
A4 = np.empty(100)
i = 0
for n in range(10,1010,10):
    G = g2(R[:n])
    A4[i] =np.mean(G)
    i = i+1
```

```python
plt.plot(N,A4)
plt.plot(N,np.ones(100)*(np.exp(1)-1))
plt.show()
```
