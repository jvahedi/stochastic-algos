```python
import numpy as np
import scipy.stats as sy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#discretation size
s = 101
#initial condition
x_0  = 1

#time discretization
t = np.linspace(0,5,s)
del_t = t[1]-t[0]

def iterate1():
    #trajectory array
    X = np.zeros(s)
    X[0]=x_0

    #array of brownian steps
    del_W = np.random.normal(0,np.sqrt(del_t),s)

    #stepping iteration through trajectory

    for j in range(1,s):
        X[j] = X[j-1]*(1-del_t + 2*del_W[j] + 2*((del_W[j])**2-del_t))
        #print([del_t, del_W[j],(-del_t + 2*del_W[j]), X[j-1]])
        
    return X

def iterate2():
    #trajectory array
    X = np.zeros(s)
    X[0]=x_0

    #array of brownian steps
    del_W = np.random.normal(0,np.sqrt(del_t),s)

    #stepping iteration through trajectory

    for j in range(1,s):
        X[j] = X[j-1]*(1-del_t + 2*del_W[j])
        #print([del_t, del_W[j],(-del_t + 2*del_W[j]), X[j-1]])
        
    return X
```

```python
n = 100
Y = np.zeros((n,s))
for i in range(n):
    Y[i] = iterate1()

P1 = np.average(Y, axis = 0)

Y = np.zeros((n,s))
for i in range(n):
    Y[i] = iterate2()
P2 = np.average(Y, axis = 0)

plt.plot(t,P1)
plt.plot(t,np.exp(-t))
#plt.plot(t,P2)
plt.show()
```
